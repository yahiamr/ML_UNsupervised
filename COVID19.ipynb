```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=2)
df['cluster'] = model.fit_predict(df.drop(columns='Country'))
```

```python
# Calculate the mean of each column by cluster
cluster_means = df.groupby('cluster').mean()
```

```python
import matplotlib.pyplot as plt

# Transpose for easier plotting
cluster_means_transposed = cluster_means.T

# Plotting
cluster_means_transposed.plot(kind='bar', figsize=(15, 10))
plt.title('Proportion of Feature per Cluster')
plt.ylabel('Proportion')
plt.xlabel('Feature')
plt.legend(title='Cluster')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Optional: Dimensionality Reduction for Visualization
pca = PCA(n_components=2)
reduced_df = pca.fit_transform(df.drop(columns=['Country', 'cluster']))

# Plotting
plt.scatter(reduced_df[:, 0], reduced_df[:, 1], c=df['cluster'])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Clusters Visualization')
plt.show()
```
